```python
import pandas as pd
import numpy as np
import re
import string
import nltk
```

```python
df = pd.read_csv('/content/test tweets.csv')
df.head()
```

```python
# Defining a function that removes pattern in the input text
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for word in r:
        input_txt = re.sub(word, "", input_txt)
    return input_txt
```

```python
# Using defined function to remove twitter handles (@user)
df['clean_tweet'] = np.vectorize(remove_pattern)(df['Tweet'], "@[\w]*")
```

```python
df.head()
```

```python
# remove emoji, punctuation and numbers 
df['clean_tweet'] = df['clean_tweet'].str.replace("[^a-zA-Z#]", " ")
df.head()
```

```python
# Import stopwords from nltk package
nltk.download('stopwords')
from nltk.stem.snowball import stopwords
stop = stopwords.words('english')
```

```python
# Iterating through each tweet, checking if words appear in stopword dictionary, remove if they do 
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

df.head()
```

```python
#Converting tweets to lowercase 
df['clean_tweet'] = df['clean_tweet'].str.lower()
df.head()
```

```python
#Importing Tokenizing function 
from nltk.tokenize import TweetTokenizer
tt = TweetTokenizer()
```

```python
#Applying Tokenizing function to dataframe

df['clean_tweet'] = df['clean_tweet'].apply(tt.tokenize)
```
